```python

#2 	Identify the columns with missing values

df = pd.read_csv('https://raw.githubusercontent.com/priyaahema/ML_DL_Resources/main/Exercises/Day%201/lung%20cancer.csv')
cv = df.columns[df.isnull().any()].tolist()
print("missing values in col:", cv)

```

```python
df.isnull().sum()
```

```python
# 3 Fill the  missing values using any one method

column_name='AGE'
df[column_name]=df[column_name].fillna(df[column_name].mean())
print(df)
```

```python
df.isnull().sum()
```

```python
# 4 Visualize the dataset using two appropriate plots

#histogram

import matplotlib.pyplot as plt
plt.figure(figsize=(5,3))
plt.hist(df['CHEST PAIN'], bins=10)
plt.xlabel('levels')
plt.ylabel('range')
plt.title('LUNG_CANCER')
plt.show()
```

```python
# scatter plot
plt.scatter(df['AGE'], df['CHRONIC DISEASE'])
plt.xlabel('age')
plt.ylabel('chronic disease')
plt.title('Scatter Plot')
plt.show()
```

```python
# 5 	Use any 2 out of 3 normalization methods to normalize two columns in the given data

#minmaxscaler
from sklearn.preprocessing import MinMaxScaler,StandardScaler
columns_to_normalize=["YELLOW_FINGERS","SHORTNESS OF BREATH"]
minmaxscaler=MinMaxScaler()
df_minmax=df.copy()
df_minmax[columns_to_normalize]=minmaxscaler.fit_transform(df[columns_to_normalize])
print("Min_Max_Scaling:")
print(df_minmax[columns_to_normalize].head())
```

```python
#standardscaler
z_score_scaler=StandardScaler()
df_z_score=df.copy()
df_z_score[columns_to_normalize]=z_score_scaler.fit_transform(df[columns_to_normalize])
print("Z_score_Scaling:")
print(df_z_score[columns_to_normalize].head())
```

```python
# 6	Use any 2 out of 5 encoding methods to encode the data

from sklearn.preprocessing import LabelEncoder
df_onehot=pd.get_dummies(df)
print("One-Hot Encoding:")
print(df_onehot.head())
```

```python
df_label=df.copy()
label_encoder=LabelEncoder()
for column in df_label.columns:
  df_label[column]=label_encoder.fit_transform(df_label[column])
  print("Label Encoding")
  print(df_label.head())
```
